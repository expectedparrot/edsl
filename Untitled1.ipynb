```python
from edsl import QuestionFreeText
q = QuestionFreeText(question_name = "name_change_thoughts", 
                     question_text = "MSNBC is thinking of changing its name to 'MS Now' - what are your thoughts?"
                    )
results = q.run()
```

```python
results.select("name_change_thoughts")
```
