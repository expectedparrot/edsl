```python
ep = FirecrawlScenario().scrape("www.expectedparrot.com")
```

```python
from edsl import QuestionNumerical
q = QuestionNumerical(question_text = "How many free credits are offerd? {{ scenario.content }}", 
                      question_name = "credits")
results = q.by(ep).run()
```
